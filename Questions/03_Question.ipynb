```python
import matplotlib.pyplot as plt
```

```python
#Create a pie chart showing the market share of mobile OS(Android, IOS, others). Then recreate the same using a horizontal bar chart and observe 
#which is easier to understand.
```

```python
labels = ['Android', 'iOS', 'Others']
market_share = [72, 26, 2]
colors = ['green', 'lightblue', 'red']
```

```python
plt.figure(figsize=(6, 6))
plt.pie(market_share, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Mobile OS Market Share (Pie Chart)')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.barh(labels, market_share, color=colors)
plt.xlabel('Market Share (%)')
plt.title('Mobile OS Market Share (Bar Chart)')
plt.show()
```
