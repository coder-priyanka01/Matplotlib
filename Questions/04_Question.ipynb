```python
import matplotlib.pyplot as plt
```

```python
#Try making a stackplot with your own weekly schedule.
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
```

```python
sleep = [8] * 7
work_study = [8, 8, 8, 8, 8, 3, 3]
free_time = [4] * 7
exercise = [1] * 7
```

```python
plt.figure(figsize=(10, 6))
plt.stackplot(days, sleep, work_study, free_time, exercise, 
              labels=['Sleep', 'Work/Study', 'Free time', 'Exercise'],
              colors=['#99ccff', '#ff9999', '#66cc66', '#ffcc99'])

plt.title('My Weekly Schedule')
plt.ylabel('Hours per Day')
plt.legend(loc='upper left')
plt.ylim(0, 24)
plt.grid(alpha=0.3)
plt.show()
```
