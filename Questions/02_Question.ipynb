```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df =  pd.read_excel("10 years of cricket data.xlsx")
```

```python
df
```

```python
Years = df["Years"]
x = np.arange(len(Years))
width = 0.25
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x - width, df['Rohit Sharma '], width, label='Rohit Sharma ', color='blue')
plt.bar(x, df['Virat Kohli'], width, label='Virat Kohli', color='green')
plt.bar(x + width, df['Sehwag'], width, label='Sehwag', color='orange')

plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Runs Comparison Over 5 Years')
plt.xticks(x, Years)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

```python
debut_runs = df.iloc[0:5][['Rohit Sharma ', 'Virat Kohli', 'Sehwag']].sum()

plt.figure(figsize=(8, 5))
plt.barh(debut_runs.index, debut_runs.values, color=['blue', 'green', 'orange'])

plt.xlabel('Total Runs')
plt.title('Total Runs in Debut 5 Years')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
```
