```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Create a plot comparing kohli, rohit sharma and sehwag across 10 years of hypothetical runs. 
#Use: labels, legends, colors, line styles, one custom style
```

```python
plt.style.use('fivethirtyeight')
```

```python
df =  pd.read_excel("10 years of cricket data.xlsx")
```

```python
df
```

```python
Years = df['Years']
Virat_Kohli_runs = df['Virat Kohli']
Rohit_Sharma_runs = df['Rohit Sharma ']
Sehwag_runs = df['Sehwag']
```

```python
print(df.columns.tolist())
```

```python
import numpy as np
x = np.arange(len(Years))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, Virat_Kohli_runs, label="Virat Kohli", color='skyblue', marker='o', linestyle='-')
plt.plot(x, Rohit_Sharma_runs, label="Rohit Sharma", color='green', marker='s', linestyle='--')
plt.plot(x, Sehwag_runs, label="Sehwag", color='red', marker='^', linestyle='-.')
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, Years)  # Show actual year values on x-axis
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
